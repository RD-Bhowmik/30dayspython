```python
num_one = 5
num_two = 4
total = num_one + num_two
multiply = num_one * num_two
divide = num_one / num_two
modulus = num_one % num_two
power = num_one ** num_two
floor_division = num_one // num_two
print(f"Addition: {total}")
print(f"Multiplication: {multiply}")
print(f"Division: {divide}")
print(f"Modulus: {modulus}")
print(f"Power: {power}")
print(f"Floor Division: {floor_division}")

radius = 30
area_of_circle = 3.14 * radius ** 2
circum_of_circle = 2 * 3.14 * radius
print(f"Area of Circle: {area_of_circle}")
print(f"Circumference of Circle: {circum_of_circle}")
print("-------------------------------------")


print("after user input")
radius = int(input("Enter radius: "))
area_of_circle = 3.14 * radius ** 2
circum_of_circle = 2 * 3.14 * radius
print(f"Area of Circle: {area_of_circle}")
print(f"Circumference of Circle: {circum_of_circle}")

```
