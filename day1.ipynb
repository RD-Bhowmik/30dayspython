```python
a = 3 
b = 4 

def addition(a, b):
    add = a + b
    print(add)
    
def sub(a, b):
    subb = a - b    
    print(subb)

def multi(a, b):
    multip = a * b
    print(multip)

def divi(a, b):
    divide = a / b
    print(divide)
    
def expo(a, b):
    exp = a ** b
    print(exp)
    
def floor_div(a, b):
    floordiv = a // b
    print(floordiv)
    

addition(a, b)
sub(a, b)
multi(a, b)
floor_div(a, b)
divi(a, b)
expo(a, b)


```

```python
import math 

integer = 2 
floating = 3.14
complex_num = 4-4j 

string = "Ronodeep Bhowmik"
boolean = True
list_data = ['Ronodeep', 'Bhowmik', 24, 'Bangladesh']
tuple_data = ('Ronodeep', 'Bhowmik', 24, 'Bangladesh')
set_data = {'Ronodeep', 'Bhowmik', 24, 'Bangladesh'}
dict_data = {
    'first_name': 'Ronodeep',
    'last_name': 'Bhowmik',
    'age': 24,
    'country': 'Bangladesh'
}
print(f"Integer: {integer}, Type: ----- {type(integer)}")
print(f"Floating Point: {floating}, Type: ----- {type(floating)}")
print(f"Complex Number: {complex_num}, Type: ----- {type(complex_num)}")  
print(f"String: {string}, Type: ----- {type(string)}")
print(f"Boolean: {boolean}, Type: ----- {type(boolean)}")
print(f"List: {list_data}, Type: ----- {type(list_data)}")
print(f"Tuple: {tuple_data}, Type: ----- {type(tuple_data)}")
print(f"Set: {set_data}, Type: ----- {type(set_data)}")
print(f"Dictionary: {dict_data}, Type: ----- {type(dict_data)}")



#Euclidean distance between two points (2, 3) and (10, 8)

p1 = (2, 3)
p2 = (10, 8)
dis = math.sqrt((p2[0] - p1[0])**2 + (p2[1] - p1[1])**2)
print(f"The Euclidean distance between points {p1} and {p2} is === {dis}")
```
